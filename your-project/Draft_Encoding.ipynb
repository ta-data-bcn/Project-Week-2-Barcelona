We are now going to clean the names of the districts so they match with the ones in the other dataframe

```python
#Defining the function to clean the string
def cleaner_str(x):
    for index,i in enumerate(x):
        if x != 'BARCELONA':
            new_i = x[3:]
        
        else:
            new_i = x
    return new_i

#Applying the variable to each row in the column Distritos
data_dist3['Distritos'] = data_dist3['Distritos'].apply(cleaner_str)
data_dist3
```
