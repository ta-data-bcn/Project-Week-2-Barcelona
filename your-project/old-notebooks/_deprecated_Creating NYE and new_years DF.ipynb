```python
import pandas as pd
import numpy as np

accidents = pd.read_csv("datasets/1.-Transportation/accidents-2017.csv")
```

```python
accidents[(accidents["Month"] == "December") & (accidents["Part of the day"].isin(["Morning","Night"]))]
NYE = accidents[(accidents["Month"] == "December") & (accidents["Day"].isin([31]))]
```

```python
accidents[(accidents["Month"] == "January") & (accidents["Part of the day"].isin(["Morning","Night"]))]
new_years = accidents[(accidents["Month"] == "January") & (accidents["Day"].isin([1]))]
```

```python
new_years_all = pd.concat([NYE, new_years], axis=1)
```
