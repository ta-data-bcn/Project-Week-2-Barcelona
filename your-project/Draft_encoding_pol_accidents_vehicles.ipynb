## Top colors of vehicles involved in accidents, grouped by the top % vehicle types

```python
# now we will get the number of vehicles involved, grouped by the top 5 types of vehicles involved in accidents, and color.

top5types = veh['Descripcio_tipus_vehicle'].value_counts().index[0:5].tolist()
print(top5types)
```

```python
# Display the vehicles with most accidents
veh_top5types = veh[veh['Descripcio_tipus_vehicle'].isin(top5types)]
veh_top5types['Descripcio_tipus_vehicle'].value_counts()
```

```python
# Finally we group by type and color, and sort them in descending order in order to get the top colors.
veh_bytypecolor = veh_top5types.groupby(['Descripcio_tipus_vehicle', 'Descripcio_color'])['Numero_expedient'].count().to_frame()
veh_bytypecolor = veh_bytypecolor.sort_values(['Descripcio_tipus_vehicle','Numero_expedient'], ascending = [False,False])

veh_bytypecolor.index.names = ['Vehicle_type', 'Color']
veh_bytypecolor.columns = ['Vehicles_involved_in_acc']

# We also add a new column as a percentage
veh_bytypecolor_perc = veh_bytypecolor.groupby(level=0).apply(lambda x: round (100 * x / x.sum(),2))
veh_bytypecolor_perc.columns = ['%Vehicles_involved_in_acc']
veh_bytypecolor = pd.concat([veh_bytypecolor,veh_bytypecolor_perc], axis =1)

# in order to get the most important data, we will filter and display only the data with more than 1% of the vehicles.
veh_bytypecolor_top = veh_bytypecolor[veh_bytypecolor['%Vehicles_involved_in_acc'] >= 1]
veh_bytypecolor_top
```

## Vehicles involved groupe by the driver license's age

```python
# let's get the vehicles implied in accidents by the age of the driver's license.
veh_license_age = veh['Antiguitat_carnet'].value_counts()

# first, we drop the unknown ones (Desconegut)
veh_license_age = veh_license_age.drop(index='Desconegut')

# now we convert the indexes type to integer
veh_license_age.index = veh_license_age.index.astype(int)
veh_license_age.index.name = 'Age_drivers_license'

# we convert the series to a dataframe, and sort by index
veh_license_age = veh_license_age.to_frame(name='Vehicles_Implied')
veh_license_age = veh_license_age.sort_index()
veh_license_age.head(30)
```

```python
# as there are a lot of years, we will group the driver license age by groups of 5 years. for that we will create a function.

def groupbyage (x):
    if x < 5:
        return 'Less than 5 years'
    elif 5 <= x < 10:
        return 'Between 5 and 10 years'
    elif 10 <= x < 15:
        return 'Between 10 and 15 years'
    elif 15 <= x < 20:
        return 'Between 15 and 20 years'
    elif 20 <= x < 25:
        return 'Between 20 and 25 years'
    elif 25 <= x < 30:
        return 'Between 25 and 30 years'
    elif 30 <= x < 35:
        return 'Between 30 and 35 years'
    elif 35 <= x < 40:
        return 'Between 35 and 40 years'
    elif 40 <= x < 45:
        return 'Between 40 and 45 years'
    elif 45 <= x < 50:
        return 'Between 45 and 50 years'
    else:
        return 'More than 50 years'

# let's group by the function created

veh_license_agegroups = veh_license_age.groupby(groupbyage).sum()
veh_license_agegroups.index.name='Drivers_license_age_group'
```

```python
# finally we reorder the rows in a logic way, from less years to more years, and display the result.
veh_license_agegroups= veh_license_agegroups.reindex(['Less than 5 years', 'Between 5 and 10 years', 'Between 10 and 15 years',
                         'Between 15 and 20 years', 'Between 20 and 25 years', 'Between 25 and 30 years',
                         'Between 30 and 35 years', 'Between 35 and 40 years', 'Between 40 and 45 years',
                         'Between 45 and 50 years', 'More than 50 years'])
veh_license_agegroups
```
