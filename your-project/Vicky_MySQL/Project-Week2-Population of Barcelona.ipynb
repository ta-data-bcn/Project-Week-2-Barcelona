### Importing Data Sets created with MySQL

```python
pop_dis_2017 = pd.read_csv('population_by_district 2017')
pop_dis_2017
```

### Piecing together the new Dataframe

```python
pop_dis17 = pop_dis_2017.rename(columns={"SUM(Number)": "Number2017"})
```

```python
pop_dis_2013 = pd.read_csv('population_by_district 2013')
```

```python
pop_dis13 = pop_dis_2013.rename(columns={"SUM(Number)": "Number2013", "District.Name" : "District.Namerem"})
```

```python
pop_dis = pd.concat([pop_dis13, pop_dis17], axis = 1).reindex(pop_dis13.index)
pop_dis
```

```python
pop_dis1 = pop_dis.drop(["Year", "District.Namerem"], axis = 1)
pop_dis1
```

#### Adding the Area as a Variable

```python
area = [4.49, 7.46, 21.35, 6.08, 20.09, 4.19, 11.96, 8.04, 6.56, 10.80]
pop_dis1['Area_km2'] = area
pop_dis1
```

### Performing the caluclations

#### Density

```python
density17 = pop_dis1.Number2017 / pop_dis1.Area_km2
density17col = list(round(density17, 2))
```

```python
pop_dis1
```

```python
density13 = pop_dis1.Number2013 / pop_dis1.Area_km2
density13col = list(round(density13, 2))
```

```python
pop_dis1['Density13'] = density13col
pop_dis1
```

#### Change

```python
Change_Num_Col = ((pop_dis1.Number2017 - pop_dis1.Number2013) / pop_dis1.Number2013) * 100
ch_num = list(round(Change_Num_Col, 2))
pop_dis1['Change.Num%'] = ch_num
pop_dis1
```

```python
Change_Dens_Col = ((pop_dis1.Density17 - pop_dis1.Density13) / pop_dis1.Density13) * 100
ch_dens = list(round(Change_Dens_Col, 2))
pop_dis1['Change.Dens%'] = ch_dens
pop_dis1
```
