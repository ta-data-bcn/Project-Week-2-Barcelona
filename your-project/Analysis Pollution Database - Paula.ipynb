```python
#Importing the Data Frame that includes information about air quality from 6 months in 2018
pollution= pd.read_csv("../datasets/2.-Urban-Environment/air_quality_2018_project.csv", index_col="Unnamed: 0").copy()
#Making a copy from the Data Set to add the proper changes and printing the head to check that everything looks good
pollution.head()
```

```python
#Checking that all the Data Types are what we expected 
pollution.info()
#The columns with hora_o3,valor_o3,hora_no2,valor_no2,hora_pm10,valor_pm10 should be integers. 
```

```python
#Cleaning the pollution dataset.Changing its columns to make more sense 

#Taking the unit from the numerical columns to make then a int datatype instead of an object datatype
pollution["nom_cabina"]=pollution["nom_cabina"].str.replace(".*(- )","")
pollution["valor_o3"]=pollution["valor_o3"].str.replace("µg/m³","")
pollution["valor_no2"]=pollution["valor_no2"].str.replace("µg/m³","")
pollution["valor_pm10"]=pollution["valor_pm10"].str.replace("µg/m³","")

# test transformed data
pollution.columns
```

```python
#Deleting columns that are not needed for our statistycal analysis
pollution1=pollution.drop(columns=["codi_dtes", "zqa","codi_eoi", "longitud", "latitud", "hora_o3", "hora_pm10", "hora_no2", "dateTime"])
#Printint the different air stations
print(pollution1["nom_cabina"].value_counts())
pollution1.head()
```

```python
#Changing name of the columns 
pollution1.columns = ["Name_Station","Air_Quality","Qualityo3","Valueo3","Qualityno2","Valueno2","Qualitypm10","Valuepm10","Generated"]
pollution1.columns
```

```python
pollution1.head(20)
```

```python
pollution1.info()
#The columns for value and generated should be numbers and not object types.
#To turn them into int we have to create a subset where NaN are either a 0 or drop them
#We are going to replace the generated date by a number too 
```

```python
print("There are",pollution1.Valueo3.isnull().sum(),"null values out of 35983")
#Almost half of the values in the values of O3 are empty. 
#Let's replace the NaN values for " "
pollution1.Valueo3 = pollution1["Valueo3"].fillna("")
pollution1.Valueo3 = pollution1["Valueo3"].str.replace("   ", "")
pollution1.Valueo3 = pollution1["Valueo3"].str.replace("--", "")
pollution1.sort_values("Valueo3")
```

```python
print("There are",pollution1.Valueno2.isnull().sum(),"null values out of 35983")
pollution1.Valueno2 = pollution1["Valueno2"].fillna("")
pollution1.Valueno2 = pollution1["Valueno2"].str.replace("   ", "")
pollution1.Valueno2 = pollution1["Valueno2"].str.replace("--", "")
pollution1.sort_values("Valueno2")
```

```python
print("There are",pollution1.Valuepm10.isnull().sum(),"null values out of 35983")
pollution1.Valuepm10 = pollution1["Valuepm10"].fillna("")
pollution1.Valuepm10 = pollution1["Valuepm10"].str.replace("   ", "")
pollution1.Valuepm10 = pollution1["Valuepm10"].str.replace("--", "")
pollution1.sort_values("Valuepm10")
```

```python
pollution_cleaned = pollution1[pollution1.Valueo3 != ""]
pollution_cleaned = pollution_cleaned[pollution_cleaned.Valueno2 != ""]
pollution_cleaned = pollution_cleaned[pollution_cleaned.Valuepm10 != ""]

pollution_cleaned.tail(20)
```

```python
pollution_cleaned.Generated.astype(str)
```

```python
def month(x):
    return x.split("/")[1]


pollution_cleaned["Month"] = pollution_cleaned.Generated.apply(month)
```

```python
#Create a function to stablish a name indicating the district
def district(row):
    if row["nom_cabina"]=="Palau Reial":
        return "Les corts"
    if row["nom_cabina"]=="Sants":
        return "Sants-Montjuïc"
    if row["nom_cabina"]=="Poblenou":
        return "Sant Martí"
    if row["nom_cabina"]=="Observ Fabra":
        return "Sarrià-Sant Gervasi"
    if row["nom_cabina"]=="Eixample":
        return "Eixample"
    if row["nom_cabina"]=="Gràcia":
        return "Gràcia"
```
