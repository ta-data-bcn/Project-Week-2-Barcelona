```python
danger_data = merged_death_rent.copy()
danger_data.rename(columns = {"Índex RFD Barcelona = 100":"Revenue_Index"},inplace = True)
#We create 4 revenue categories for the negihborhoods based on their revenue levels
danger_data["Category"] = ["Very poor" if e < 63 
                                          else "Poor" if e < 79
                                          else "Low middle class" if e < 100
                                          else "High middle class" if e < 126
                                          else "Rich" if e < 159
                                          else "Very rich"
                                          for e in danger_data["Revenue_Index"]]
danger_data.rename(columns = {'Number':'Number_Deaths','Nombre':'Life_Expectancy'},inplace = True)
danger_data.drop(['Revenue_Index','Year','District.Code','Neighborhood.Code'],axis=1,inplace = True)
danger_data[danger_data["Neighborhood.Name"] == "les Tres Torres"]
```

```python
#We group the deaths and population by age group and neighborhood
deaths_by_age = danger_data.groupby(["Neighborhood.Name","Age"],as_index=False).aggregate({'District.Name':'first','Number_Deaths':'sum','Població':'first','Life_Expectancy':'mean','Category':'first'})
```

```python
#We compute the life expectancy for each category
life_expectancies = deaths_by_age.groupby("Category")[["Life_Expectancy"]].mean().sort_values(by = "Life_Expectancy",ascending =  False)
```

```python
#We compute the population and neighborhoods by category:
life_expectancies_refined = deaths_by_age[["Neighborhood.Name","Població","Life_Expectancy","Category"]].groupby(["Neighborhood.Name"],as_index = False).first()
life_expectancies_refined["Neighborhoods"] = 1
summary_table = life_expectancies_refined.groupby("Category")["Població","Neighborhoods"].sum()
summary_table["% of population"] = summary_table["Població"] / summary_table["Població"].sum()
summary_table["% of population"] = pd.Series(["{0:.2f}%".format(val*100) for val in summary_table["% of population"]],index = summary_table.index)
summary_table.rename(columns = {"Població":"Population"},inplace=True)
summary_table
```

```python
#We compute a more precise life expectancy by category
life_expectancies_refined["Ponderated_pop"] = life_expectancies_refined["Població"] * life_expectancies_refined["Life_Expectancy"]
life_exp = life_expectancies_refined.groupby("Category")[["Població","Ponderated_pop"]].sum()
life_exp["Life Expectancy"] = round(life_exp["Ponderated_pop"]/life_exp["Població"],2)
life_exp.drop(['Població','Ponderated_pop'],axis=1,inplace=True)
life_exp.sort_values(by = 'Life Expectancy',ascending = False)
```
