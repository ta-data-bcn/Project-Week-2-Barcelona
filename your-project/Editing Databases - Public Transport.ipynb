```python
#Importing the Data Frame that includes information about air quality
pollution=pd.read_csv("../datasets/2.-Urban-Environment/air_quality_2018_project2.csv", index_col="Unnamed: 0")
pollution.head()
```

```python
#Cleaning the pollution dataset. We are taking the Barcelona - info from the neighborhoods to have it in the same format as the info about neighborhoods in our public transportation dataframe
#Taking the unit from the numerical columns to make then a int datatype instead of an object datatype
pollution["nom_cabina"]=pollution["nom_cabina"].str.replace(".*(- )","")
pollution["valor_o3"]=pollution["valor_o3"].str.replace(" µg/m³","")
pollution["valor_no2"]=pollution["valor_no2"].str.replace(" µg/m³","")
pollution["valor_pm10"]=pollution["valor_pm10"].str.replace(" µg/m³","")
# test transformed data
pollution.columns
```

```python
#Deleting columns that are not needed for our statistycal analysis
pollution1=pollution.drop(columns=["codi_dtes", "zqa","codi_eoi", "longitud", "latitud", "hora_o3", "hora_pm10", "hora_no2", "qualitat_o3", "qualitat_pm10", "qualitat_no2", "dateTime"])
#Printint the different air stations
print(pollution1["nom_cabina"].value_counts())
#Printing the unique names of the neighborhoods
neighborhood_list=list(bcn_stops["Neighborhood.Name"].unique())
district_list=list(bcn_stops["District.Name"].unique())
#Printing the unique names of the Districts 
print(district_list)
```

```python
#We have each Air Station to a different district

#Gràcia corresponds to Gràcia
#Ciutadella corresponds to Poblenou
#Sants corresponds to Sants-Montjuïc
#Vall Hebron corresponds to Horta-Guinardó
#Poblenou corresponds to Sant Martí
#Eixample corresponds to Eixample
#Observ Fabra corresponds to Sarrià-Sant Gervasi
```

```python
#Create a function to stablish a name indicating the district
def district(row):
    if row["nom_cabina"]=="Palau Reial":
        return "Les corts"
    if row["nom_cabina"]=="Sants":
        return "Sants-Montjuïc"
    if row["nom_cabina"]=="Poblenou":
        return "Sant Martí"
    if row["nom_cabina"]=="Observ Fabra":
        return "Sarrià-Sant Gervasi"
    if row["nom_cabina"]=="Eixample":
        return "Eixample"
    if row["nom_cabina"]=="Gràcia":
        return "Gràcia"
```

```python
pollution1["District"]= pollution1.apply(district, axis=1)
```

```python
pollution1.head()
```

```python
pollution1.info()
```

```python
pollution1.District = pollution1.District.fillna("")

pollution_districts = pollution1[pollution1.District != ""]
```

```python
pollution_districts.info()
```

```python
pollution_districts.valor_no2 = pollution_districts.valor_no2.fillna("")
pollution_districts.valor_no2 = pollution_districts.valor_no2.str.replace("--", "")
pollution_districts.valor_o3 = pollution_districts.valor_o3.fillna("")
pollution_districts.valor_o3 = pollution_districts.valor_o3.str.replace("--", "")
pollution_districts.valor_pm10 = pollution_districts.valor_pm10.fillna("")
pollution_districts.valor_pm10 = pollution_districts.valor_pm10.str.replace("--", "")
```

```python
districts_o3=pollution_districts.loc[pollution_districts["valor_o3"]!="", ["District", "valor_o3"]]
districts_o3.reset_index
districts_no2=pollution_districts.loc[pollution_districts["valor_no2"]!="", ["District", "valor_no2"]]
districts_no2.reset_index
districts_pm10=pollution_districts.loc[pollution_districts["valor_pm10"]!="", ["District", "valor_pm10"]]
districts_pm10.reset_index
```

```python
def only_digits(x):
    if x.isdigit():
        return int(x)
    else:
        return np.NaN
```

```python
districts_pm10["valor_pm10"] = districts_pm10["valor_pm10"].apply(only_digits)
```

```python
districts_o3["valor_o3"] = districts_o3["valor_o3"].apply(only_digits)
districts_no2["valor_no2"] = districts_no2["valor_no2"].apply(only_digits)
```

```python
grouped_districts_pm10=districts_pm10.groupby("District")["valor_pm10"].mean()
grouped_districts_o3=districts_o3.groupby("District")["valor_o3"].mean()
grouped_districts_no2=districts_no2.groupby("District")["valor_no2"].mean()
print(grouped_districts_no2)
print(grouped_districts_o3)
print(grouped_districts_pm10)
```

```python
district_names=bcn_stops["District.Name"].unique()
#distric_names=district_names.to_numpy()
#district_names=pd.Series(district_names)
print(type(district_names))
grouped_stops=bcn_stops.groupby("District.Name")["District.Name"].count()
grouped_stop=grouped_stops.to_numpy()
print((type(grouped_stop)))
```

```python
stops_per_district=pd.DataFrame({"district":district_names, "stops":grouped_stop}, columns=["district", "stops"])
stops_per_district
```

```python
mean_pm10_district= pd.DataFrame(districts_pm10.groupby("District")["valor_pm10"].mean())
mean_pm10_district["index"]=[0,1,2,3,4,5]
mean_pm10_district.set_index("index")
mean_pm10_district["district"]=["Eixample", "Gràcia", "Les Corts", "Sant Martí", "Sants-Montjuïc", "Sarrià-Sant Gervasi"]
mean_pm10_district=mean_pm10_district[["district","valor_pm10", "index"]]
mean_pm10_district.set_index("index")
```

```python
mean_o3_district= pd.DataFrame(districts_o3.groupby("District")["valor_o3"].mean())
mean_o3_district["index"]=[0,1,2,3,4,5]
mean_o3_district["district"]=["Eixample", "Gràcia", "Les Corts", "Sant Martí", "Sants-Montjuïc", "Sarrià-Sant Gervasi"]
mean_o3_district=mean_o3_district[["district","valor_o3", "index"]]
mean_o3_district.set_index("index")
```

```python
mean_no2_district= pd.DataFrame(districts_no2.groupby("District")["valor_no2"].mean())
mean_no2_district["index"]=[0,1,2,3,4,5]
mean_no2_district["district"]=["Eixample", "Gràcia", "Les Corts", "Sant Martí", "Sants-Montjuïc", "Sarrià-Sant Gervasi"]
mean_no2_district=mean_no2_district[["district","valor_no2", "index"]]
mean_no2_district.set_index("index")
```

```python
final_database=pd.merge(mean_no2_district, mean_o3_district, how="inner", on="district")
final_database=final_database[["district", "valor_no2", "valor_o3"]]
final_database=pd.merge(final_database, mean_pm10_district, how="inner", on="district")
final_database=final_database[["district", "valor_no2", "valor_o3", "valor_pm10"]]
final_database=pd.merge(final_database, stops_per_district, how="inner", on="district")
final_database
```
