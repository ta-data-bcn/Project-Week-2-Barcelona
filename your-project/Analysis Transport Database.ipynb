```python
import pandas as pd
import numpy as np
```

```python
#Importing the Data sets of public transportation in Barcelona
bus_stops = pd.read_csv("../datasets/1.-Transportation/bus-stops.csv",index_col = 0)
public_stops = pd.read_csv("../datasets/1.-Transportation/public-transport.csv", index_col = 0)
```

```python
#Printing first rows to check if the columns are equal 
print(bus_stops.head())
```

```python
#Printing first rows to check if the columns are equal 
print(public_stops.head())
```

```python
#Changing the title and creating the DataFrame of Bus.Stop column to Station to have it equal in both datasets
bus_stops = bus_stops.rename(columns = {"Bus.Stop":"Station"})
bus_stops.head()
```

```python
#By typing info into the dataframes we can understand data types and how many null values we have
bus_stops.info()
```

```python
#There are some null values on the neighborhoods that we need to figure out why. 
public_stops.info()
```

```python
#Merging both Data Frames together to have a full data set with all the information about the different public transportations stops in Barcelona
bcn_stops = public_stops.append(bus_stops)
```

```python
bcn_stops
```

```python
#Calculating the number of Neighborhoods
bcn_stops["Neighborhood.Name"].value_counts()
#There are 73 different neighborhoods. 
```

```python
#Checking if there are any duplicates
bcn_stops[bcn_stops.Station.duplicated()]
#There are around 1500 rows that are duplicates, that demonstrates that some stations have different transports 
```

```python
#Saving the Dataframe to a csv 
bcn_stops.to_csv("../datasets/1.-Transportation/bcn_stops.csv")
```

```python
bcn_stops["Neighborhood.Name"].isnull()
external_stops=bcn_stops.loc[bcn_stops['Neighborhood.Name'].isnull()]
external_stops #External stops are from outside Barcelona!
```

```python
#We are editing the data frame we had by selecting only the rows that are not null in the neighborhood column
bcn_stops = bcn_stops.loc[bcn_stops['Neighborhood.Name'].notna()]
bcn_stops
```

```python
stops_per_district_and_transport = pd.DataFrame(bcn_stops.groupby(['District.Name', 'Transport']).size())
stops_per_district_and_transport
```

```python
stops_per_district_and_transport.to_csv("../datasets/1.-Transportation/stops_per_district_and_transport.csv")
```
