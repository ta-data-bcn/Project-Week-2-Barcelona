```python
import pandas as pd 

data = pd.read_csv('/Users/jordijordanagarcia/Desktop/Ironhack/WEEK2/Project-Week-2-Barcelona/datasets/3.-Population/immigrants-by-nationality.csv', sep=';')
```

```python
dataset = data[['Year', 'District Name', 'Neighborhood Name','Nationality', 'Number']]
```

```python
fdataset = dataset.loc[(dataset['Nationality'] != 'Spain') & (dataset['Nationality'] !='No information')]
```

```python
data2015 = fdataset.groupby(['Year', 'Neighborhood Name', 'Nationality']).max().loc[2015]
```

```python
data2015.sort_values('Number', ascending=False).head(5)
```

```python
data2016 = fdataset.groupby(['Year', 'Neighborhood Name', 'Nationality']).max().loc[2016]
```

```python
data2016.sort_values('Number', ascending=False).head(5)
```
